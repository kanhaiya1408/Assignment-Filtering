[cite_start]**Question 6: Show employees who are from 'Delhi' or 'Mumbai' and earn more than 55,000.** [cite: 44]

[cite_start]**Question 7: Display all employees except those from the 'HR' department.** [cite: 45]

[cite_start]**Question 8: Get all employees hired between 2019 and 2022, ordered by HireDate (oldest first).** [cite: 46]

```python
SELECT * FROM Employees
WHERE HireDate BETWEEN '2019-01-01' AND '2022-12-31'
ORDER BY HireDate ASC;
```
